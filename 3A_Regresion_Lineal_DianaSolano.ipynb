# Error cuadrático medio

<br><br>
$$ MSE = \frac{1}{N}\sum_{i=1}^{N}(real_i - \ estimado_i)^2 $$
<br><br>

<img style="width:50%" src="https://evidenciasenpediatria.es/files/43-224-RUTA/46_figura_1.jpg" >


<div style="font-family:courier new;margin-top:40px;margin-bottom:40px">
El Error Cuadrático Medio es el criterio de evaluación más usado para problemas de regresión.  

Se usa sobre todo cuando usamos aprendizaje automático supervisado. Para cada dato histórico podremos indicar el resultado correcto.

Vamos a ver como se calcula<br><br>
<div style=";font-size:30px">
$$error \ cuadr\acute{a}tico = (real\ - \ estimado)^2$$
</div>

Error real menos error estimado, tiene lógica ... pero; ¿por qué al cuadrado? <br><br>

- Simple, no queremos valores negativos, para tomar de esa manera, el 0, como valor perfecto.  

Otro punto a tener en cuenta, es que tranquilamente podríamos usar los valores absolutos de los errores, o sea, simplemente quitándoles el signo. Pero esto resultaría en una función no derivable, algo que nos jugará en contra a futuro, apliquemos algoritmos de optimización muy efectivos como el gradiente descendente.
<br><br>

Ya sabemos cómo calcular el error de cada uno de los puntos, por lo que podremos calcular entonces, el error medio.  
Para ello sumamos todos los errores, para luego dividir el resultado por la cantidad de puntos o errores que acabamos de sumar.
<div style=";font-size:30px"> 
$$MSE = \frac{1}{M}\sum_{i=1}^{M}(real_i - \ estimado_i)^2$$ 
</div>

No te preocupes, no te abandonaré aquí:<br><br>
<b>MSE: </b> Error Cuadrático Medio (Mean Squared Error) <br><br>
<b>M: </b> Número total de puntos o predicciones<br><br>

$
\begin{align}
\sum_{i=1}^{M}:
\end{align}
$
</b>   Sumatorio (Letra Griega Sigma)
<div>


# Raíz cuadrada del error cuadrático 

<div style="font-family:courier new;margin-top:40px;margin-bottom:40px">
<div style="font-size:30px">
$$RMSE = \sqrt{\frac{1}{M}\sum_{i=1}^{M}(real_i - \ estimado_i)^2}$$
<br><br>
</div>

El error cuadrático, es preciso, pero algo engañoso, ya que si vemos en una 
estimación de precios que nuestro sistema tiene un error cuadrático medio de 
1 millón de dólares, tenemos que tener presente de que sumamos potencias de 
2, en otras palabras sumamos números que previamente los elevamos al 
cuadrado! <br><br>

Por lo que el error real, será la raiz cuadrada de ese valor, al cual, por sus 
siglas en inglés lo denominamos RMSE.<br><br>

Para resumir, a la hora de visualizar puede que el RMSE nos de una sensación 
mas realista, pero a la hora de computar, el MSE tiene una carga computacional 
mucho menor, por eso es que es más usado
</div>
